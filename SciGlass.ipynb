```python
import pubchempy as pcp


def get_smiles(chemical_name):
    """Fetch SMILES using PubChem, handling errors properly."""
    try:
        comp = pcp.get_compounds(chemical_name, "name")
        return comp[0].isomeric_smiles if comp else None
    except Exception:
        return None


# Convert polymer names into possible monomer equivalents
monomer_names = {
    "Poly(benzyl acrylate)": "Benzyl acrylate",
    "Poly(4-biphenyl acrylate)": "4-Biphenyl acrylate",
    "Poly(butyl acrylate)": "Butyl acrylate",
    "Poly(sec-butyl acrylate)": "Sec-butyl acrylate",
    "Poly(2-tertbutylphenyl acrylate)": "2-Tertbutylphenyl acrylate",
}

# Try to fetch SMILES for monomers instead of full polymer names
monomer_smiles = {poly: get_smiles(mono) for poly, mono in monomer_names.items()}

# Convert to a DataFrame for easy handling
import pandas as pd

df_smiles = pd.DataFrame.from_dict(monomer_smiles, orient="index", columns=["SMILES"])
print(df_smiles)
```
